<h1 style="color: green;">Monitoring datasets preparation</h1>
<p>
EvidentlyAI monitoring is very specific as to how the data should be prepared.<br> 
In this section we prepare the datasets for classifier and regressor models monitoring.<br><br>
The following tasks will be accomplished:<br>
<ul>
<li><b>Extract classifier and regressor feaures datasets from homeshopping.csv</b></li><br>
<li>Create classifier reference dataset, this is the <b>benchmark dataset</b>;<br>it consists of all the records in homeshopping for the feature in question</li>
<li>Make predictions with the reference dataset and append them back to the dataset</li>
<li>Export the reference dataset</li><br>

<li style="color: green;">Create classifier current no drift dataset;<br>this consists of random sampling from the reference dataset</li>
<li style="color: green;">Make predictions with the classifier current no drift dataset and append them back to the dataset</li>
<li style="color: green;">Export the classifier current no drift dataset</li><br>

<li style="color: red;">Create classifier current dataset with drift;<br>this consists of adding random noise to the current no drift dataset</li>
<li style="color: red;">Make predictions with the classifier current dataset with drift and append them back to the dataset</li>
<li style="color: red;">Export the classifier current dataset with drift</li><br><br>

    
    

<li>Create regressor reference dataset, this is the <b>benchmark dataset</b>;<br>it consists of all the records in homeshopping for the feature in question</li>
<li>Make predictions with the reference dataset and append them back to the dataset</li>
<li>Export the reference dataset</li><br>

<li style="color: green;">Create regressor current no drift dataset;<br>this consists of random sampling from the reference dataset</li>
<li style="color: green;">Make predictions with the regressor current no drift dataset and append them back to the dataset</li>
<li style="color: green;">Export the regressor current no drift dataset</li><br>

<li style="color: red;">Create regressor current dataset with drift;<br>this consists of adding random noise to the current no drift dataset</li>
<li style="color: red;">Make predictions with the regressor current dataset with drift and append them back to the dataset</li>
<li style="color: red;">Export the regressor current dataset with drift</li>

</ul>
</p>

```python
import pandas as pd
import numpy as np
import joblib

import copy

import make_predictions
```

<h1 style="color: green;">Restricting data to classifier model features</h1>

```python
# This includes the classifier target (Trips_response_lt_5)
features_classifier = ['Date_diff', 'Nbr_items_per_wk', 'Expenditure_per_wk',
                        'Total_Exp_wk_perc', 'AVG_exp_item_per_wk', 'Bread_wk',
                        'Cooked_meats_wk', 'Raw_meats_wk', 'Eating_out_wk', 'Snacks_wk',
                        'Drinks_wk', 'Vegetables_wk', 'Fruit_wk', 'Cooking_base_wk',
                        'Dairy_produce_wk', 'Breakfast_wk', 'Transport_wk',
                        'Tech_and_services_wk', 'Cosmetics_and_selfcare_wk',
                        'House_and_kitchen_wk','Trips_response_lt_5']
```

```python
classifier_dt = copy.deepcopy(raw0)
```

```python
classifier_dt = classifier_dt[features_classifier]
classifier_dt.head()
```

```python
classifier_dt1 = copy.deepcopy(classifier_dt)
```

```python
# extracting the target, Trips_response_lt_5
classifier_target = classifier_dt1.Trips_response_lt_5

# dropping Trips_response_lt_5 from classifier_dt
classifier_dt1.drop(['Trips_response_lt_5'], axis=1, inplace=True)
classifier_dt1.shape
```

<h5 style="color: green;">Make predictions</h5>

```python
classifier_predictions_prob, classifier_predictions_binarized = \
make_predictions.predict_classifier(classifier_dt1)
classifier_predictions_binarized[:10]
```

```python
# create the reference data
classifier_reference = copy.deepcopy(classifier_dt)

# append the predictions and rename data to classifier_dt
classifier_reference['prediction_label'] = classifier_predictions_binarized
classifier_reference['prediction_prob'] = classifier_predictions_prob
```

<h3 style="color: green;">Current dataset with drift</h3>

```python
classifier_drift = copy.deepcopy(Classifier_curr_dt_no_drift)
```

```python
classifier_drift.head()
```

```python
classifier_wk_list = ['Date_diff', 'Nbr_items_per_wk', 'Expenditure_per_wk',
                        'AVG_exp_item_per_wk', 'Bread_wk',
                        'Cooked_meats_wk', 'Raw_meats_wk', 'Eating_out_wk', 'Snacks_wk',
                        'Drinks_wk', 'Vegetables_wk', 'Fruit_wk', 'Cooking_base_wk',
                        'Dairy_produce_wk', 'Breakfast_wk', 'Transport_wk',
                        'Tech_and_services_wk', 'Cosmetics_and_selfcare_wk',
                        'House_and_kitchen_wk']

classifier_perc_list = ['Total_Exp_wk_perc']
```

<h5 style="color: green;">Inflating values to mimic drift</h5>

```python
# Adding random noise to the non percentage features
# In general all the values are being inflated
mu, sigma = 10,6 
for item in classifier_wk_list:
    classifier_drift['{}'.format(item)] = classifier_drift['{}'.format(item)] + \
    abs(np.random.normal(mu, sigma,len(classifier_drift)))
classifier_drift[classifier_wk_list].head()
```

```python
# Adding random noise to the weekly percentages features
mu, sigma = 0,0.1
for item in classifier_perc_list:
    classifier_drift['{}'.format(item)] = classifier_drift['{}'.format(item)] + \
    abs(np.random.normal(mu, sigma,len(classifier_drift)))
classifier_drift[classifier_perc_list].head()
```

```python
# Although this exceeds 1, there is no need to cap this at 1
# as the idea is to demonstrate drift in the data
classifier_drift.Total_Exp_wk_perc.max()
```

```python
classifier_drift.columns
```

```python
classifier_drift1 = copy.deepcopy(classifier_drift)
```

```python
classifier_drift.shape
```

<h5 style="color: green;">Extracting the classifier target</h5>

```python
classifier_drift_target = classifier_drift1.Trips_response_lt_5

# dropping Trips_response_lt_5 from classifier_dt
classifier_drift1.drop(['Trips_response_lt_5'], axis=1, inplace=True)
classifier_drift1.shape
```

<h5 style="color: green;">Make predictions</h5>

```python
classifier_drift_predictions_prob, classifier_drift_predictions_binarized=\
make_predictions.predict_classifier(classifier_drift1)
classifier_drift_predictions_prob[:10]
```

<h5 style="color: green;">Append predictions to current dataset no drift</h5>

```python
# classifier_drift['prediction'] = classifier_drift_predictions

# append the predictions and rename data to classifier_dt
classifier_drift['prediction_label'] = classifier_drift_predictions_binarized
classifier_drift['prediction_prob'] = classifier_drift_predictions_prob
```

```python
print(classifier_drift.shape)

classifier_drift.head()
```

```python
# Export classifier_drift
classifier_drift.to_csv("../2. Data/classifier_current_dt_drift.csv",index=False)
```
